# Bias voltage = -0.5V

```python
amps = get_amplitudes("attenuation -- circle fibre holder", 'max')
```

```python
distances = [165]
d = 165
for i in range(amps[2] - 1):
    d += 53.4
    distances.append(round(d))
distances = np.array(distances) / 1000
distances
```

```python
len(distances)
```

```python
opt = scipy.optimize.curve_fit(lambda t,a,b: a*np.exp(b*t),  distances[8:],  amps[0][8:])
a_1, b_1 = opt[0]
a_1, b_1
```

```python
amps[0]
```

```python
summary_attenuation([distances], [amps], ['1.2mm'], ['orange'])
```

# Bias voltage = -0.7V

```python
amps_07 = get_amplitudes("test 2", 'max')
```

```python
distances_07 = [165]
d = 165
for i in range(amps_07[2] - 1):
    d += (53.4 * 4)
    distances_07.append(round(d))
distances_07 = np.array(distances_07) / 1000
distances_07
```

```python
opt = scipy.optimize.curve_fit(lambda t,a,b,c: a*np.exp(b*t) + c,  distances_07,  amps_07[0])
a_1_07, b_1_07, c_1_07 = opt[0]
a_1_07, b_1_07, c_1_07
```

```python
summary_attenuation([distances_07], [amps_07], ['1.2'], ['purple'])
```
