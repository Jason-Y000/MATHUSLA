# 2-4 Photon Test

Here we apply a very low bias voltage (Vbias = -1.08 V) and use the laser pulser as the trigger to be able to probe regions of 2-4 photons without worrying about dark counts getting coupled into the data. Done with BCF 9299A fibres of barying diameters

```python
distances_ul = [327]
d= 327
for i in range(7):
    d += 51.05 * 6
    distances_ul.append(round(d))
distances_ul = np.array(distances_ul) / 1000
print(distances_ul)
```

```python
ultralow =  Attenuation(distances_ul,"ultralowintensity/1.2mm", channels=['1', '4'], maxx='min')
```

```python
summary_attn(distances_ul, [ultralow.amplitudes()], ["1.2"], ['orange'])
```

```python
ultralow18 =  Attenuation(distances_ul,"ultralowintensity/1.8mm", channels=['1', '4'], maxx='min')
```

```python
summary_attn(distances_ul, [ultralow18.amplitudes()], ["1.8"], ['hotpink'])
```

```python
ultralow20 =  Attenuation(distances_ul,"ultralowintensity/2.0mm", channels=['1', '4'], maxx='min')
```

```python
summary_attn(distances_ul, [ultralow20.amplitudes()], ["2.0"], ['forestgreen'])
```

```python
summary_attn(distances_ul, [ultralow.amplitudes(), ultralow18.amplitudes(), ultralow20.amplitudes()], ["1.2", "1.8", "2.0"], colors=["orange","hotpink", "forestgreen"])
```
