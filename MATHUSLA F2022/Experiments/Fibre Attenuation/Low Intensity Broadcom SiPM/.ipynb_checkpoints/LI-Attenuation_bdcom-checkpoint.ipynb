# Low Intensity Region Attenuation Measurements

This notebook is for the attenuation data processing in the low intensity region (<20 photons).  We process the data using the library and plot 

- Summary of Channel 1
- Summary of Channel 4
- 


bias voltage -0.88V (lowest voltage that gives Ch4 a signal from 3m of fibre away for 1.2mm fibre)

```python
import sys        
sys.path.append('/Users/lanabozanic/MATHUSLA') # Change this to the dir where mathusladataprocessing.py is.     

from mathusladataprocessing import *
```

```python
distances = [327]
d= 327
for i in range(14):
    d += 51.05 * 3
    distances.append(round(d))
distances = np.array(distances) / 1000
print(distances)
```

# 1.2mm Fibre

```python
a12 = Attenuation(distances, "Final data/1.2mm LIR", channels=['1', '4'], maxx='min')
amps_12 = a12.amplitudes()
```

```python
print("Channel 1 Summary")
print("Min. photons Ch1: {:.3f}-{:.3f} photons".format(abs(max(amps_12['1']) * 1000 / 7.40) - 2.38, abs(max(amps_12['1']) * 1000/ 7.40) + 2.38))
print("Max. photons Ch1: {:.3f}-{:.3f} photons\n".format(abs(min(amps_12['1']) * 1000/ 7.40)- 2.38, abs(min(amps_12['1']) * 1000/ 7.40)+ 2.38))


print("Channel 4 Summary")
print("Min. photons Ch4: {:.3f}-{:.3f} photons".format(abs(max(amps_12['4']) * 1000/ 7.40)- 2.38, abs(max(amps_12['4']) * 1000/ 7.40)+ 2.38))
print("Max. photons Ch4: {:.3f}-{:.3f} photons".format(abs(min(amps_12['4']) * 1000/ 7.40)- 2.38, abs(min(amps_12['4']) * 1000/ 7.40)+ 2.38))
```

```python
summary_attn(distances, [amps_12], ["1.2"], ['hotpink'])
```
