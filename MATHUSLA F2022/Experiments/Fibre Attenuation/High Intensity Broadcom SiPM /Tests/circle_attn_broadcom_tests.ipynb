# Dataset 2

```python
distances_2 = [190]
d = 190
for i in range(4):
    d += 53.4 * 4
    distances_2.append(round(d))
distances_2 = np.array(distances_2) / 1000
distances_2
```

We lower the trigger for the second attempt, and find the 'true' distribution of a pulse, which include exponentially decaying noise near 0. When the amplitude of the signal is larger enough, it is simple enough to filter the noise out but this can pose a problem for low intensity measurements whose trigger is unbiased.

```python
bdcom2 = Attenuation(distances_2, "broadcom test 2", channels=['1','4'],maxx='min')
means2, stds2 =bdcom2.histogram(bins=20)
```

```python
summary_attn(distances_2, [bdcom2.amplitudes()], ['1.2'], colors=['blue'])
```
