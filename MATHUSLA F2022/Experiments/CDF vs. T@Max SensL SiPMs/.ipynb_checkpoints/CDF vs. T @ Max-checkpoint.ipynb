## Summary of the data:

```python
stds_tr = []
stds_meas = []
for i in range(len(td_trace_all)):
    stds_tr.append(np.std(td_trace_all[i]))
    stds_meas.append(np.std(td_meas_all[i]))
    
plt.scatter(delay_distances, stds_tr, label = 'Constant Fraction (70%)')
plt.scatter(delay_distances, stds_meas, label = 'Leading Edge')
plt.ylim(min(stds_tr) * 0.5, max(stds_meas) * 1.2)
plt.legend(fontsize = 13, frameon=True)
plt.xlabel('Distance from Ch1 to Pulse (cm)')
plt.ylabel('Standard Deviation (s)')
plt.title('Constant Fraction vs Leading Edge Timing Methods')
plt.show()
```

```python
means_tr = []
means_meas = []

for i in range(len(td_trace_all)):
    means_tr.append(np.mean(td_trace_all[i]))
    means_meas.append(np.mean(td_meas_all[i]))
    
plt.scatter(delay_distances, means_tr, label = 'Constant Fraction (70%)')
plt.scatter(delay_distances, means_meas, label = 'Leading Edge')
plt.ylim(-16e-9,-2e-9 )
plt.legend(fontsize = 13, frameon=True)
plt.xlabel('Distance from Ch1 to Pulse (cm)')
plt.ylabel('Delay Time')
plt.title('Constant Fraction vs Leading Edge Timing Methods')
plt.show()
```
