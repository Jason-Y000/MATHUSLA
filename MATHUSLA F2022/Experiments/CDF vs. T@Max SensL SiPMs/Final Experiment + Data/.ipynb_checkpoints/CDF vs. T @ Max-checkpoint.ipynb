# CDF vs. T@Max Timing Methods

This notebook compares the timing resolution achieved by either using the constant fraction discrimination method, or the T@Max method.

Read more about these methods here: https://ns.ph.liv.ac.uk/~ajb/ukgs_nis/pre-course-material/lec5-03.pdf

## Summary of the data:

```python
stds_tr = []
stds_meas = []
for i in range(len(td_trace_all)):
    stds_tr.append(np.std(td_trace_all[i]))
    stds_meas.append(np.std(td_meas_all[i]))
    
plt.scatter(delay_distances, stds_tr, label = "Constant Fraction 100%")
plt.scatter(delay_distances, stds_meas, label = 'Leading Edge')
plt.ylim(min(stds_tr) * 0.5, max(stds_meas) * 1.2)
plt.legend(fontsize = 13, frameon=True)
plt.xlabel('Distance from Ch1 to Pulse (mm)')
plt.ylabel('Standard Deviation in Time Delay (s)')
plt.title('Constant Fraction vs t@Max: standard devs.')
plt.show()
```

```python
means_tr = []
means_meas = []

for i in range(len(td_trace_all)):
    means_tr.append(np.mean(td_trace_all[i]))
    means_meas.append(np.mean(td_meas_all[i]))
    
plt.scatter(delay_distances, means_tr, label = "Constant Fraction 100%")
plt.scatter(delay_distances, means_meas, label = 'Leading Edge')
plt.ylim(-16e-9,-2e-9 )
plt.legend(fontsize = 13, frameon=True)
plt.xlabel('Distance from Ch1 to Pulse (mm)')
plt.ylabel('Time Delay (s)')
plt.title('Constant Fraction vs t@Max: mean time delay')
plt.show()
```

```python
fractions = np.linspace(0,1, 11)
fractions[10] = 0.99999
fractions = fractions[2:]
fractions
```

```python
plt.figure(figsize=(8,5.5))
path = 'sensL sipm delays 205 mm/SNSL_sipm_trace_205mm'
colors = ["red", "chocolate", "orange", "gold", "yellow", "yellowgreen", "green","turquoise", "blue", "mediumpurple", "purple"]
h_shift = 0.3e-8

for i in range(len(fractions)):
    
    td = get_timing_delays_trace(path, 20, fractions[i], False)
    mu, std = norm.fit(np.array(td) + h_shift)
    plt.hist(np.array(td) + h_shift, density=True, alpha=0.6, color=colors[i], bins=10, label="{:0.2}, std = {:0.3} ns".format(fractions[i], std * 1e9))
    
    # gaussian fit plot

    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2,color=colors[i])
    plt.xlim(-1.5e-8,0.2e-8)
    print("std for {:.1f}% CDF: {:.3f} ns".format(fractions[i] * 100, std * 1e9))
    
    h_shift += 0.1e-8

plt.xlabel("Delay (s)")
plt.ylabel("Frequency")
plt.title("CDF with different fractions of amplitude, 205mm Delay Distance\n")
plt.legend()
plt.show()
```
