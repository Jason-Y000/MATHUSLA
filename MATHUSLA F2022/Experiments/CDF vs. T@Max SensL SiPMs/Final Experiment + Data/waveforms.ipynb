```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math as math
import scipy
from scipy.stats import norm
import os, glob
```

```python
# Graph the pulse
def graph_pulses(pathname):
    counter = 0
    
    # Grab the data
    #data.columns

    csv_files = glob.glob(os.path.join(pathname, "*.csv"))


    # loop over the list of csv files
    for f in csv_files:
        if counter == 5:
            return
        fig, ax = plt.subplots(figsize=(8,6))
        # read the csv file
        data = pd.read_csv(f, skiprows=1)

        # print the location and filename

        p1 = np.array(data['1 (VOLT)'])
        p2 = np.array(data['4 (VOLT)'])
    
        p1_filter = scipy.ndimage.gaussian_filter(p1, sigma = 20)
        p2_filter = scipy.ndimage.gaussian_filter(p2, sigma = 20)
        
        time = np.array(data['Time (s)'])
        
        plt.xlabel("time (s)", fontsize=14)
        plt.ylabel("amplitude (V)", fontsize=14)
        plt.title("Pulse graph, raw and filtered", fontsize=14)
        plt.plot(time, p1,color='plum', label='Channel 1, Raw')
        plt.plot(time, p2,color='cornflowerblue',label='Channel 4, Raw')
        plt.plot(time, p1_filter,color='indigo',linestyle='--', label='Channel 1, Filtered', linewidth=5)
        plt.plot(time, p2_filter,color='royalblue',linestyle='--', label='Channel 4, Filtered',linewidth=5)
        
        plt.legend()
        plt.show()
        
        counter +=1

```

```python
graph_pulses("sensL sipm delays 205 mm/SNSL_sipm_trace_20_f")
```
