Plotting the amplitude histogram

```python
path = "Darkcount Trace SenslAmp"
```

```python
maxes=[]
csv_files = glob.glob(os.path.join(path, "*.csv"))
# loop over the list of csv files
for f in csv_files:
    # read the csv file
    data = pd.read_csv(f, skiprows=1)
    maxes.append(max(data['1 (VOLT)']))
```
