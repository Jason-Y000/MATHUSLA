If we fit a single gaussian (and use a low binning), we can't see the single photoelectron peaks.

```python
a1 = Attenuation([], "external trigger (Laser Pulser)", channels=['4'], maxx='min')
m20, std20, = a1.histogram(bins=20)
```

But if we lower the binning, and fit a gaussian mixture, we can see the photo electron peaks. They get more clear at lower voltages!

```python
means, stds = a1.histogram(single_fit=False,bins=130)
```

To find the amplitude of a single photo electron, we take the means of each gaussian found, and take the difference. In the graph above, each red dot is a photoelectron peak, that determines how many gaussians we will fit in the gaussian mixture.

```python
pes=[]
for meanz in means['4']:
    pe=0
    #print(means)
    m = sorted(meanz, key = lambda x:float(x))
    prev = m[0]
    for mean in m[1:]:
        pe += (mean-prev) * 1000
        prev = mean
    
    pe=pe/(len(m)-1)
    print("avg pe:",pe)
    
    pes.append(pe)
    
print("final avg. pe: (noise included):", sum(pes)/len(pes))
```

```python
s=[]
for ss in stds['4']:
    for i in ss:
        s.append(np.sqrt(i[0][0]))

sum(s)/len(s) * 1000 
```

So 1 photoelectron has an energy of $ \approx 7.48 \pm 2.24$ photons (crude approximation)
